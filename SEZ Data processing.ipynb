## Introduction

This notebook presents an in-depth analysis of Pakistan's economic indicators and the impact of launching CPEC on the GDP of Pakistan. The analysis aims to explore the complex relationship between various factors and economic growth, providing valuable insights for policymakers and stakeholders.

- **China_Imports**: China's total imports from the country in USD.

# **Economic Analysis: PCA to DiD Regression**

### Addressing Multicollinearity
Due to the multicollinearity issue observed in the regression analysis, steps were taken to mitigate it. Specifically, related variables were grouped into indexes to reduce multicollinearity.

### Principal Component Analysis (PCA)
Principal Component Analysis (PCA) was employed to create these composite variables.
- **Economic Index**: This index includes variables related to economic indicators such as GDP, imports, exports, net trade, etc.
- **Social Index**: Variables related to social indicators like unemployment rate, population, urban population percentage, etc., were grouped into this index.
- **Industrial Index**: This index comprises variables associated with industrial indicators, specifically manufacturing and industry value added.


```python
from sklearn.decomposition import PCA

# Define groups of related variables
Economic_Index = ['GDP ($)','Imports ($)','Exports ($)', 'Net_Trade ($)', 'Dollar_Rate', 'FDI_Inflows_(BOP)', 'Total_Reserves_($)','D.S_External_debt']
Social_Index = ['Unemployment Rate (%)', 'Population', 'Urban_Population_%','Access_to_Electricty (%)','Labour_F.P.R']
Industrial_Index = ['Manafacturing (V.A $)', 'Industry (V.A $)']

# Combine related variables into separate DataFrames
economic_indicators_df = df[Economic_Index]
social_df = df[Social_Index]
industrial_df = df[Industrial_Index]

# Apply PCA to each group of related variables
pca_economic_indicators = PCA(n_components=1)
pca_social = PCA(n_components=1)
pca_industrial = PCA(n_components=1)

economic_indicators_pca = pca_economic_indicators.fit_transform(economic_indicators_df)
social_pca = pca_social.fit_transform(social_df)
industrial_pca = pca_industrial.fit_transform(industrial_df)

# Add the principal components to your DataFrame as composite variables/index
df['Economic_Indicators_Index'] = economic_indicators_pca
df['Social_Index'] = social_pca
df['Industrial_Index'] = industrial_pca
```

## Evolution of Modeling Strategy

Initially, I constructed composite indexes, namely Economic_Indicators_Index, Social_Index, and Industrial_Index, through Principal Component Analysis (PCA) to encapsulate groups of related variables. However, upon analyzing the results, it became apparent that these indexes did not yield satisfactory outcomes in explaining the variations in GDP.

### Refinement of Approach

To address this limitation, I adopted a refined method. Firstly, I identified positively and negatively correlated variables based on their relationship with GDP. Positively correlated variables, such as imports, exports, population, etc., and negatively correlated variables, such as net trade, balance of payments, etc., were segregated into two distinct groups.

```python
import pandas as pd
from sklearn.decomposition import PCA

# Define positively and negatively correlated variables
positively_correlated = ['Imports ($)','Exports ($)', 'Population', 'Access_to_Electricty (%)', 
                         'FDI_Inflows_(BOP)', 'Total_Reserves_($)', 'Dollar_Rate', 
                          'Manafacturing (V.A $)', 'Industry (V.A $)', 
                         'Labour_F.P.R', 'Urban_Population_%', 'China_Exports', 'China_Imports']

negatively_correlated = ['Net_Trade ($)', 'China_Trade_Balance']

# Separate positively and negatively correlated variables
positively_df = df[positively_correlated]
negatively_df = df[negatively_correlated]

# Apply PCA to positively correlated variables
pca_positively = PCA(n_components=1)
positively_pca = pca_positively.fit_transform(positively_df)

# Apply PCA to negatively correlated variables
pca_negatively = PCA(n_components=1)
negatively_pca = pca_negatively.fit_transform(negatively_df)

# Add the principal components to your DataFrame as composite variables/index
df['Positively_Index'] = positively_pca
df['Negatively_Index'] = negatively_pca
```

```python
import statsmodels.api as sm

# Define your independent variables (indices) and the third variable
X = df[['Positively_Index', 'Negatively_Index', 'D.S_External_debt','Balance_Of_Payments($)','CPEC_launched','SEZ_Act_Lauched']]
y = df['GDP ($)']

# Add a constant term
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

### Explanation of Difference-in-Differences (DiD) Analysis

Difference-in-Differences (DiD) analysis is a statistical technique used to estimate the causal effect of a treatment or intervention by comparing the changes in outcomes between a treatment group and a control group before and after the treatment is implemented. Here's how the DiD analysis is applied to this data:

#### Step 1: Define Treatment and Control Groups
- In your dataset, the treatment group consists of observations where the CPEC (China–Pakistan Economic Corridor) project was launched (CPEC_launched == 1).
- The control group consists of observations where the CPEC project was not launched (CPEC_launched == 0).

#### Step 2: Create Dummy Variables
- **Treatment Dummy (treatment_dummy)**: This dummy variable takes the value of 1 for observations in the treatment group and 0 otherwise. It identifies whether an observation is in the treatment group.
- **Post-Intervention Dummy (post_intervention_dummy)**: This dummy variable takes the value of 1 for observations occurring after the intervention period (post-2015 in this case) and 0 otherwise. It identifies whether an observation is in the post-intervention period.

#### Step 3: Create the Interaction Term
- The interaction term (did) is the product of the treatment dummy and the post-intervention dummy. It captures the combined effect of the treatment and the post-intervention period.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Assuming df is your DataFrame with the relevant data

# Define treatment and control groups
treatment_group = df[df['CPEC_launched'] == 1]
control_group = df[df['CPEC_launched'] == 0]

# Create dummy variables for treatment and post-intervention period
df['treatment_dummy'] = (df['CPEC_launched'] == 1).astype(int)
df['post_intervention_dummy'] = (df['Year'] >= 2015).astype(int)

# Create the interaction term for DiD
df['did'] = df['treatment_dummy'] * df['post_intervention_dummy']

# Define the regression model
# Define the relevant independent variables and dependent variable
X = df[['treatment_dummy', 'post_intervention_dummy', 'did']]
y = df['GDP ($)']

# Add a constant term
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

In the regression results, the coefficient estimates represent the average change in GDP for each unit change in the respective independent variables.

- The coefficient for the `const` term (intercept) represents the average GDP value when all other independent variables are zero.
- The coefficient for `treatment_dummy` represents the average difference in GDP between the treatment group (CPEC launched) and the control group (CPEC not launched), holding other variables constant. However, it appears to be statistically insignificant (p-value > 0.05), suggesting that there is no significant difference in GDP between the treatment and control groups before the intervention.
- The coefficient for `post_intervention_dummy` represents the average change in GDP after the intervention (CPEC launch), compared to before the intervention, holding other variables constant. This coefficient is statistically significant (p-value < 0.05), indicating a significant increase in GDP after the intervention.
- The coefficient for `did` represents the difference-in-differences effect, which captures the differential change in GDP between the treatment and control groups after the intervention. However, it also appears to be statistically insignificant (p-value > 0.05), suggesting that there is no significant differential change in GDP between the treatment and control groups after the intervention.

Overall, the model's R-squared value (0.736) indicates that approximately 73.6% of the variance in GDP is explained by the independent variables included in the model. However, the presence of multicollinearity or singular design matrix issues may affect the reliability of the coefficient estimates and the overall model fit. Further diagnostics may be needed to address these issues.

#### Explanation of Revised Difference-in-Differences (DiD) Analysis

After conducting the initial DiD analysis and observing unsatisfactory results, a revised DiD analysis is performed using a different set of variables. Here's how the revised DiD analysis is conducted:

##### Creating a Composite Index Using PCA
- A set of relevant variables is identified, including imports, exports, FDI inflows, total reserves, exchange rates, external debt, manufacturing value added, industry value added, balance of payments, China's exports and imports, and the number of SEZs launched.
- These relevant variables are combined into a composite index using Principal Component Analysis (PCA). PCA helps reduce the dimensionality of the dataset while retaining the most important information.

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import statsmodels.api as sm

# Assuming your DataFrame is named df and contains the relevant variables

# Combine relevant variables into an index using PCA
relevant_variables = ['Imports ($)','Exports ($)','FDI_Inflows_(BOP)', 'Total_Reserves_($)',
       'Dollar_Rate', 'D.S_External_debt', 'Manafacturing (V.A $)',
       'Industry (V.A $)', 'Balance_Of_Payments($)',
       'China_Exports', 'China_Imports','No_Of_SEZs_Launched']
df_relevant = df[relevant_variables]

# Perform PCA to create the composite index
pca = PCA(n_components=1)
composite_index = pca.fit_transform(df_relevant)

# Add the composite index to the DataFrame
df['Composite_Index'] = composite_index

# Define treatment and control groups
df['Treatment_Group'] = np.where(df['Year'] >= 2015, 1, 0)  # Assuming CPEC launched in 2015
df['Post_Intervention'] = np.where(df['Year'] >= 2015, 1, 0)

# Create the DiD variable
df['DiD'] = df['Treatment_Group'] * df['Post_Intervention']

# Run the DiD regression model
X = df[['Composite_Index', 'DiD']]
X = sm.add_constant(X)  # Add constant term
y = df['GDP ($)']

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

Interpretation:

- **R-squared**: The R-squared value of 0.969 indicates that approximately 96.9% of the variation in GDP ($) can be explained by the independent variables (Composite Index and DiD) included in the model. This suggests that the model fits the data quite well.

- **Adjusted R-squared**: The adjusted R-squared value of 0.965 is similar to the R-squared value but adjusts for the number of independent variables in the model. It penalizes the inclusion of unnecessary variables and is a more conservative measure of model fit.

- **F-statistic**: The F-statistic of 263.2 is associated with a very low p-value (1.62e-13), indicating that the overall regression model is statistically significant. This suggests that at least one of the independent variables has a significant effect on the dependent variable (GDP ($)).

- **Coefficients**:
  - **Constant (Intercept)**: The constant coefficient represents the expected value of GDP when all independent variables are zero. In this case, it is 2.338e+11 (approximately 233.8 billion), suggesting a high baseline level of GDP ($).
  - **Composite_Index**: The coefficient of 2.5286 indicates that for a one-unit increase in the Composite Index, GDP ($) is expected to increase by approximately 2.5286 units, holding other variables constant.
  - **DiD (Difference-in-Differences)**: The coefficient of 3.876e+10 suggests that the treatment effect captured by the DiD variable leads to an additional increase in GDP ($) of approximately 38.76 billion.

- **Significance Levels (P>|t|)**: The p-values associated with each coefficient are very low, indicating that both the Composite Index and DiD variable are statistically significant predictors of GDP ($).

- **Multicollinearity**: The condition number is large, indicating potential multicollinearity issues among the independent variables. This suggests that there might be strong correlations between the Composite Index and the DiD variable or other numerical problems in the model.

```python
import matplotlib.pyplot as plt

# Plotting the predicted values of GDP for treatment and control groups
plt.figure(figsize=(10, 6))

# Plotting the actual GDP values
plt.scatter(df['Year'], df['GDP ($)'], color='gray', label='Actual GDP')

# Plotting the predicted values for treatment group
predicted_treatment = results.predict(X) + 1.418e+11
plt.plot(df['Year'], predicted_treatment, color='blue', label='Predicted GDP (Treatment)')

# Plotting the predicted values for control group
predicted_control = results.predict(X) - 1.813e+10
plt.plot(df['Year'], predicted_control, color='red', linestyle='--', label='Predicted GDP (Control)')

# Adding vertical lines to indicate the intervention (CPEC launch)
plt.axvline(x=2015, color='green', linestyle='--', label='CPEC Launch')

plt.xlabel('Year')
plt.ylabel('GDP ($)')
plt.title('Difference-in-Differences (DiD) Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

Based on the DiD model results, it appears that CPEC had a statistically significant impact on Pakistan's economy. The coefficient associated with the DiD variable is positive and statistically significant (p-value < 0.05), indicating that there was a positive change in GDP after the launch of CPEC. Therefore, we can conclude that CPEC had a positive impact on Pakistan's economy.
